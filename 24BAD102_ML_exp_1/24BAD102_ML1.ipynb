# Scenario 1


```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("carrie1/ecommerce-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(path + "/data.csv", encoding = "ISO-8859-1")
```

```python
print("Head Data:")
df.head()
```

```python
print("Tail Data:")
df.tail()
```

```python
print("Info Data:")
df.info()
```

```python
print("Describe Data:")
df.describe()
```

```python
print(df.isnull().sum())
```

```python
product_sales = df.groupby("Description")["Quantity"].sum().head(10)
```

```python
product_sales.plot(kind="bar")
plt.title("Top Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```

```python
product_sales.plot(kind="line")
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```

# Scenario 2

```python
path2 = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")
```

```python
hospital_df = pd.read_csv(path2 + "/diabetes.csv")
```

```python
print(hospital_df.head())
print(hospital_df.info())
print(hospital_df.describe())
print()
print(hospital_df.isnull().sum())
```

```python
plt.hist(hospital_df["Glucose"], bins=20)
plt.title("Glucose Level Distribution")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
print()

sns.boxplot(x=hospital_df["Age"])
plt.title("Age Distribution")
plt.show()
```

# Scenario 3

```python
path3 = kagglehub.dataset_download("harishkumardatalab/housing-price-prediction")

housing_df = pd.read_csv(path3 + "/Housing.csv")
```

```python
print(housing_df.head())
print(housing_df.info())

print()

print(housing_df.isnull().sum())
```

```python
plt.scatter(housing_df["area"], housing_df["price"])
plt.title("Area vs House Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
print()

corr = housing_df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("imakash3011/customer-personality-analysis")

print("Path to dataset files:", path)
```

```python
path4 = kagglehub.dataset_download("imakash3011/customer-personality-analysis")

bank_df = pd.read_csv(path4 + "/marketing_campaign.csv", sep="\t")

print(bank_df.head())
print(bank_df.info())
print()

print(bank_df.isnull().sum())

bank_df["Age"] = 2026 - bank_df["Year_Birth"]

bank_df["Age"].plot(kind="hist", bins=20)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print()

sns.boxplot(x=bank_df["Income"])
plt.title("Income Distribution")
plt.show()
```
