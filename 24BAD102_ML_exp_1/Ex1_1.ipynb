```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("carrie1/ecommerce-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(path + "/data.csv", encoding = "ISO-8859-1")
```

```python
print("Head Data:")
df.head()
```

```python
print("Tail Data:")
df.tail()
```

```python
print("Info Data:")
df.info()
```

```python
print("Describe Data:")
df.describe()
```

```python
print(df.isnull().sum())
```

```python
product_sales = df.groupby("Description")["Quantity"].sum().head(10)
```

```python
product_sales.plot(kind="bar")
plt.title("Top Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```

```python
product_sales.plot(kind="line")
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```
